**The IT test calculation**
<br>
GAUSS [source](https://github.com/fokep/Sanso_et_al-ICSS/blob/40d14f93a1289a88781f6f75d416937aca890fd4/icss/icss.src#L312)

```python
def itt(data):
    """Calculates the centered and normalized cumulative sums of squares
    test value and position for use with the ICSS algorithm by
    Inclan & Tiao, 1994, along with plot array
    
    Input:
        data    -- 1D numpy array with time series data
    
    Output:
        Array containing
            [0] Dk_plot -- plot values
            [1] Dk_max  -- IT - the test value (supremum of |Dk_plot|)
            [2] k_star  -- position of Dk_max in array
        
    """
    # Precur
    Ck = np.cumsum(data**2)              # make 1d-array with cumsums of data^2
    CT = Ck[-1]                          # grab the last value off Ck
    T = len(data)                        # get the length of the time-series
    ks = np.arange(1, (T+1))             # make 1d-array with index from (1:T)
    
    # Main
    Dk = (Ck / CT) - (ks / T) 
    
    # Output
    Dk_plot = np.sqrt(T/2) * Dk          # For plotting
    Dk_max  = np.amax(abs(Dk_plot))      # For For test against critical value
    k_star  = np.argmax(abs(Dk))         # For slicing
    
    return Dk_plot, Dk_max, k_star
```

**Call of the ITT function**
<br>
The test value and the position of it printed below

```python
result = itt(data)

print('Test value: {}'.format(result[1], result[2]))
print('Position: {}'.format(result[2]))
```
